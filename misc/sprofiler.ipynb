```python
%load_ext autoreload
%autoreload 2

import sprofiler as sp
from time import sleep
```

```python
!rm profiler.log
pr = sp.Profiler(verbose=2)

pr.start('program')
print('Code outside loop')
sleep(1)

for _ in range(10):
    pr.start('loop')
    print('Code in loop')
    sleep(1)
    pr.stop('loop')
pr.stop('program')

pr.report()

!cat profiler.log
```

```python
!rm profiler.log
pr = sp.Profiler(verbose=2)


@sp.time_func
def myprint():
    print('Code in loop')
    sleep(1)
        
@sp.time_func
def myfunc():
    for _ in range(10):
        myprint()

pr.start('program')
print('Code outside loop')
sleep(1)


myfunc()
        
pr.stop('program')

pr.report()

!cat profiler.log
```

```python
pr = sp.Profiler()

@pr.time_func
def sleep_1s():
    print('Code in loop')
    sleep(1)
    
@pr.time_func
def my_func():
    for _ in range(10):
        sleep_1s()
        
my_func()
pr.report()
```
