```python
%load_ext autoreload
%autoreload 2

%matplotlib inline
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../setigen/")
import setigen as stg
from astropy import units as u
import blimpy as bl

import numpy as np
```

```python
frame = stg.Frame(fchans=512,
                  tchans=16,
                  df=2.7939677238464355*u.Hz,
                  dt=18.25361108*u.s,
                  fch1=6095.214842353016*u.MHz)

frame.add_noise_from_obs(share_index=True)

signal = frame.add_constant_signal(f_start=frame.get_frequency(128),
                          drift_rate=0*u.Hz/u.s,
                          level=frame.get_intensity(snr=10),
                          width=10*u.Hz,
                          f_profile_type='gaussian')
frame.render()
```

```python
frame = stg.Frame(fchans=1024,
                  tchans=32,
                  df=2.7939677238464355*u.Hz,
                  dt=18.25361108*u.s,
                  fch1=6095.214842353016*u.MHz)

frame.add_noise_from_obs()

signal = frame.add_constant_signal(f_start=frame.get_frequency(128),
                          drift_rate=0*u.Hz/u.s,
                          level=frame.get_intensity(snr=0),
                          width=10*u.Hz,
                          f_profile_type='gaussian')
frame.render()
plt.show()

plt.hist(frame.get_data().flatten(), bins=100)
plt.show()

frame.get_data().mean(), frame.get_data().std()
```
